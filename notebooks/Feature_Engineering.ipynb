```python
import pandas as pd

df = pd.read_csv('../data/raw_churn_data.csv')

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df = df.dropna()
```

```python
#Feature 1: Charge per tenure
```

```python
df['charge_per_tenure'] = df['TotalCharges'] / (df['tenure'] + 1)
```

```python
#Feature 2: Avg monthly charge change (proxy)
```

```python
df['avg_charge_diff'] = df['MonthlyCharges'] - (df['TotalCharges'] / (df['tenure'] + 1))
```

```python
#Feature 3: Number of services used
```

```python
services = ['PhoneService', 'InternetService', 'OnlineSecurity',
            'OnlineBackup', 'DeviceProtection', 'TechSupport',
            'StreamingTV', 'StreamingMovies']

df['num_services'] = df[services].apply(lambda x: sum(x == 'Yes'), axis=1)
```

```python
#Feature 4: High-risk flag
```

```python
df['high_risk'] = (
    (df['SeniorCitizen'] == 1) &
    (df['Partner'] == 'No') &
    (df['Dependents'] == 'No')
).astype(int)
```

```python
#Feature 5: Contract encoding(ordinal logic)
```

```python
contract_map = {
    'Month-to-month': 0,
    'One year': 1,
    'Two year': 2
}

df['contract_type_encoded'] = df['Contract'].map(contract_map)
```

```python
#Feature 6: Tenure group
```

```python
df['tenure_group'] = pd.cut(df['tenure'],
                           bins=[0,12,24,48,72],
                           labels=['0-1yr','1-2yr','2-4yr','4-6yr'])
```

```python
#CLV (Customer Lifetime Value - simplified)
```

```python
df['CLV'] = df['MonthlyCharges'] * df['tenure']
```

```python
#Encode Target Variable
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df.to_csv('../data/processed_data.csv', index=False)
```

## Feature Engineering Summary:
- Created charge-based behavioral features
- Built service engagement indicator
- Engineered demographic risk persona
- Added contract ordinal encoding
- Created CLV for business decision layer
